## 1.2 t-SNE visualization of high-dimensional data

- t-SNE:t-distributed stochastic neighbor embedding
- This maximizes the distance between observations in 2 dimensional space that are most different in high dimensional space. Thus, Observations similar to one another will appear closer.
- t-SNE doesn't work for non numeric data. so remove categorical vars or one-hot encode them.
- We can however, put down categorical var info post plotting like below. Here, we see what it means to scale down high dim data to 2 dim data. There is a logical distribution to the clusters which makes sense and validates t-SNE method.

### 2.2.1 Finding a good variance threshold

```python
# Creating data for the next steps
head_df = ansur_all[['headbreadth','headcircumference','headlength','tragiontopofhead']]
```

```python
head_df.boxplot()
plt.show()
```

```python
# Normalize the data
normalized_df = head_df / head_df.mean()

normalized_df.boxplot()
plt.show()
print(normalized_df.var())

# in our example, the variance in 1.0e-03 or higher. 
# However, in DataCamp's df, 2 additional features were present that had variance of 3.231707e-27, 1.002552e-08
# Assuming we want to remove those 2 feautres, we would then select the variance threshold to be at least 0.001
# that way, anything lower will be removed from the datafrrame
```

```python
from sklearn.feature_selection import VarianceThreshold

# Create a VarianceThreshold feature selector
sel = VarianceThreshold(threshold=0.001)

# Fit the selector to normalized head_df
sel.fit(head_df / head_df.mean())

# Create a boolean mask
mask = sel.get_support()

# Apply the mask to create a reduced dataframe
reduced_df = head_df.loc[:, mask]

print("Dimensionality reduced from {} to {}.".format(head_df.shape[1], reduced_df.shape[1]))

# the dim was reduced from 6 to 4 in the DataCamp example
# However, our DF looks a bit different and there is no reduction per se.
# Think of this as an illustration
```

## 2.4 Removing highly correlated features

- There are times when some features are highly correlated. (like below).
- After analysing and making sure that the correaltion is linear, we could consider keeping only 1 of the values. Doing so could prevent the models from overfitting to the meaningless differences between the 2 models.
- Obviously there are some caveats we should be aware of:
    - We should do this only if we have good understanding of data
    - We should do this only if we think we won't loose information in doing so.
    - The correlations should be linear in nature. Remember Anscombe's quartets as an example with same correlation but different relation
    - This is a brute force method and we could also go for feature extraction instead.
    

<img src = './Images/DR-HC.png' width = 400 align = "left">  

```python
# Calculate the correlation matrix and take the absolute value
corr_matrix = ansur_all.corr().abs()

# Create a True/False mask and apply it
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)

# List column names of highly correlated features (r > 0.95)
to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.95)]

# Drop the features in the to_drop list
reduced_df = ansur_all.drop(to_drop, axis=1)

print("The reduced dataframe has {} columns.".format(reduced_df.shape[1]))
```

# 3. Feature selection II, selecting for model accuracy

- Another way of featuree selection is assessing a feature's impact on the model performance.
- We can use something like Recursive Feature Elimination. This will help us assess the impact of removing/keeping a feature on model accuracy


```python
X_p = PimaIndians.drop('test',axis=1)
y_p = PimaIndians['test']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X_p, y_p, test_size = 0.2, random_state=42)

# Creating scaler object
scaler = StandardScaler()

# Create the classifier: logreg
lr = LogisticRegression()

# Fit the scaler on the training features and transform these in one go
X_train_std = scaler.fit_transform(X_train)

# Fit the logistic regression model on the scaled training data
lr.fit(X_train_std, y_train)

# Scale the test features
X_test_std = scaler.transform(X_test)

# Predict diabetes presence on the scaled test set
y_pred = lr.predict(X_test_std)

# Prints accuracy metrics and feature coefficients
print("{0:.1%} accuracy on test set.".format(accuracy_score(y_test, y_pred))) 
print(dict(zip(X.columns, abs(lr.coef_[0]).round(2))))
```

## 3.2 Manual Recursive Feature Elimination

- Below is how we get a feature and their coeff. We can then manually remove the feature that has the lowest coeff.
- Such features might or might not have any impact on the accuracy but we need to test it to say conclusively.

```python
# Performs a 25-75% train test split
X_train, X_test, y_train, y_test = train_test_split(X_p, y_p, test_size=0.25, random_state=0)

# Scales features and fits the logistic regression model
lr.fit(scaler.fit_transform(X_train), y_train)

# Calculates the accuracy on the test set and prints coefficients
acc = accuracy_score(y_test, lr.predict(scaler.transform(X_test)))
print("{0:.1%} accuracy on test set.".format(acc)) 
print(dict(zip(X_p.columns, abs(lr.coef_[0]).round(2))))
```

## 3.3 Automatic Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE

# Create the RFE with a LogisticRegression estimator and 3 features to select
rfe = RFE(estimator=LogisticRegression(), n_features_to_select=3, verbose=1)

# Fits the eliminator to the data
rfe.fit(X_train, y_train)

# Print the features and their ranking (high = dropped early on)
print(dict(zip(X_p.columns, rfe.ranking_)))

# Print the features that are not eliminated
print(X_p.columns[rfe.support_])

# Calculates the test set accuracy
acc = accuracy_score(y_test, rfe.predict(X_test))
print("{0:.1%} accuracy on test set.".format(acc))
```

## 3.5 Regularized linear regression

```python
ansur_lasso_X = ansur_male.drop(['Branch',
 'Component',
 'Gender',
 'weight_kg',
 'stature_m',
 'BMI',
 'BMI_class',
 'Height_class'],axis=1)
ansur_lasso_y = ansur_male['BMI']
```

```python
# Import Lasso
from sklearn.linear_model import Lasso

# Set the test size to 30% to get a 70-30% train test split
X_train, X_test, y_train, y_test = train_test_split(ansur_lasso_X, ansur_lasso_y, test_size=0.3, random_state=0)

# Fit the scaler on the training features and transform these in one go
X_train_std = scaler.fit_transform(X_train)

# Create the Lasso model
la = Lasso()

# Fit it to the standardized training data
la.fit(X_train_std,y_train)
```

```python
# Transform the test set with the pre-fitted scaler
X_test_std = scaler.transform(X_test)

# Calculate the coefficient of determination (R squared) on X_test_std
r_squared = la.score(X_test_std, y_test)
print("The model can predict {0:.1%} of the variance in the test set.".format(r_squared))

# Create a list that has True values when coefficients equal 0
zero_coef = la.coef_ == 0

# Calculate how many features have a zero coefficient
n_ignored = sum(zero_coef)
print("The model has ignored {} out of {} features.".format(n_ignored, len(la.coef_)))

# We can predict almost 85% of the variance in the BMI value using just 9 out of 91 of the features. 
# The R^2 could be higher though.
```

```python
## Adjusting regularization strength

# Find the highest alpha value with R-squared above 98%
la = Lasso(alpha=0.1, random_state=0)

# Fits the model and calculates performance stats
la.fit(X_train_std, y_train)
r_squared = la.score(X_test_std, y_test)
n_ignored_features = sum(la.coef_ == 0)

# Print peformance stats 
print("The model can predict {0:.1%} of the variance in the test set.".format(r_squared))
print("{} out of {} features were ignored.".format(n_ignored_features, len(la.coef_)))
```

## 3.6 Combining feature selectors

```python
ansur_male_sub_X = ansur_male[['acromialheight',  'axillaheight',  'bideltoidbreadth',  'buttockcircumference',  'buttockkneelength',  'buttockpopliteallength',  'cervicaleheight',  'chestcircumference',  'chestheight',  'earprotrusion',  'footbreadthhorizontal',  'forearmcircumferenceflexed',  'handlength',  'headbreadth',  'heelbreadth',  'hipbreadth',  'iliocristaleheight',  'interscyeii',  'lateralfemoralepicondyleheight',  'lateralmalleolusheight',  'neckcircumferencebase',  'radialestylionlength',  'shouldercircumference',  'shoulderelbowlength',  'sleeveoutseam',  'thighcircumference',  'thighclearance',  'verticaltrunkcircumferenceusa',  'waistcircumference',  'waistdepth',  'wristheight',  'BMI' ]][:1000]
ansur_male_sub_y = ansur_male[['bicepscircumferenceflexed']][:1000]

# Set the test size to 30% to get a 70-30% train test split
X_train, X_test, y_train, y_test = train_test_split(ansur_male_sub_X, ansur_male_sub_y, test_size=0.3, random_state=0)

# Fit the scaler on the training features and transform these in one go
X_train = scaler.fit_transform(X_train)
```
