# Feature engineering

## ## Removing IDs

```python
# Storing the IDs in a separate DF
train_df_IDs = train_df['ID']
test_df_IDs = test_df['ID']

# Dropping the columns
train_df.drop(['ID'], axis=1, inplace=True)
test_df.drop(['ID'], axis=1, inplace=True)

# Checking the shape of both DFs
print(train_df.shape) 
print(test_df.shape)
```

## Removing Outliers

- For the time being we will remove the outlier with more than 80M in sales.

```python
train_df = train_df[train_df['SalesInMillions']<50].reset_index(drop=True)
```

## Missing Values

```python
# Separating Predictor and Labels
gaming = train_df.drop("SalesInMillions",axis=1)
gaming_labels = train_df['SalesInMillions']
```

```python
# calc total missing values
total_series = gaming.isnull().sum().sort_values(ascending=False)

# calc percentages
perc_series = (gaming.isnull().sum()/gaming.isnull().count()).sort_values(ascending = False)

# concatenating total values and percentages
missing_data = pd.concat([total_series, perc_series*100], axis=1, keys=['Total #', 'Percent'])

# Looking at top 20 entries
missing_data
```

- Seems we have no missing values! Yay!

## One Hot encoding categorical features + Scaling numeric features

```python
gaming.head()
```

```python
# Numeric columns to be transformed
num_cols = ['CRITICS_POINTS','USER_POINTS']

# Categorical columns to be transformed
cat_cols = ['CONSOLE','CATEGORY','PUBLISHER','RATING']
```

### Categorical pipeline

```python
# Importing necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Creating pipeline for categorical variables with missing value should be "None"
cat_pipeline = Pipeline([
        ('encoder', OneHotEncoder(sparse=False,handle_unknown='ignore'))
    ])
```

### Numeric Pipeline

```python
# Importing necessary libraries
from sklearn.preprocessing import StandardScaler

# Creating numerical pipeline
num_pipeline = Pipeline([
        ('std_scaler', StandardScaler()), # Scale the numeric features
    ])
```

### Full pipeline

```python
# Importing ColumnTransformer
from sklearn.compose import ColumnTransformer

# Creating full pipeline to process numeric and categorical features
full_pipeline = ColumnTransformer(transformers=[
        ("num", num_pipeline, num_cols),
        ("cat", cat_pipeline, cat_cols)
    ],remainder='passthrough')

# Creating prepared data by passing the training set without labels
transf = full_pipeline.fit(gaming)

gaming_clean = transf.transform(gaming)
```

```python
# check the shape
gaming_clean.shape
```

```python
# Cleaning the test data 
test_prepared = transf.transform(test_df)
```

```python
# Checking Test data's shape
test_prepared.shape
```

# Training Models

```python
# Defining CV Score function
def cv_score(score):
    rmse = np.sqrt(-score) # -score because we are using "neg_mean_squared_error" as our metric
    return (rmse)
```

## RandomForestRegressor

```python
%%time

# Import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error as MSE

# Instantiate rf
rf = RandomForestRegressor(random_state = 42)

# Creating Parameter grid for GridSearch CV
params_rf = {
    'n_estimators': [100,500,1000], # No of trees
    'max_depth': [5,10,15], # maximum depth to explore
    'min_samples_split':[5,10,15], # minimum samples required for split
    'min_samples_leaf':[5,10,15], # minimum samples required at leaf
    'max_features': ['log2', 'auto', 'sqrt'] # number of features for the best split
}

# Instantiate grid_rf
grid_rf = GridSearchCV(estimator = rf, # regressor we want to use
                       param_grid = params_rf, # Hyperparameter space
                       scoring ='neg_mean_squared_error', # MSE will be performance metric
                       cv = 3, # #of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_rf.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best RandomForestRegressor model: {:.2f}'.format(cv_score(grid_rf.best_score_)))
```

```python
# Store the best model 
best_model_RF = grid_rf.best_estimator_
```

```python
best_model_RF
```

## Gradient Boosting


```python
%%time

# Importing the necessary package
from sklearn.ensemble import GradientBoostingRegressor

# Instantiate the Gradient Boosting Regressor
gbr = GradientBoostingRegressor(subsample = 0.9,
                                max_features = 0.75,
                                random_state = 42,
                                warm_start = True,
                                learning_rate= 0.01)

# Creating Parameter grid for GridSearch CV
params_gbr = {
    'n_estimators': [1000,4000,8000],
    'max_depth': [2,5,8], 
    'min_samples_split':[5,15],
    'min_samples_leaf':[5,15],
    'max_features': ['log2', 'auto', 'sqrt']
}

# Instantiate grid search using GradientBoostingRegressor
grid_gbr = GridSearchCV(estimator = gbr, # regressor we want to use
                       param_grid = params_gbr, # Hyperparameter space
                       scoring ='neg_mean_squared_error',
                       cv = 3, # No of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_gbr.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best GradientBoostingRegressor model: {:.2f}'.format(cv_score(grid_gbr.best_score_)))
```

```python
# Store the best model 
best_model_GBR = grid_gbr.best_estimator_
```

```python
best_model_GBR
```

## Ridge Regression


```python
%%time

from sklearn.linear_model import Ridge

# Instantiate the Ridge Regressor
ridge = Ridge(random_state=42)

# Creating Parameter grid for GridSearch CV
params_ridge = {
    'alpha': [1e-15, 1e-10, 1e-8, 9e-4, 7e-4, 5e-4, 3e-4, 1e-4, 1e-3, 5e-2, 1e-2, 0.1, 0.3, 1, 3, 5, 10, 15, 18, 20, 30, 50, 75, 100],
    'solver': ['auto','saga','sag','cholesky']
}

# Instantiate grid search using GradientBoostingRegressor
grid_ridge = GridSearchCV(estimator = ridge, # regressor we want to use
                       param_grid = params_ridge, # Hyperparameter space
                       scoring ='neg_mean_squared_error',
                       cv = 3, # No of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_ridge.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best RidgeRegressor model: {:.2f}'.format(cv_score(grid_ridge.best_score_)))
```

```python
# Store the best model 
best_model_ridge = grid_ridge.best_estimator_
```

```python
best_model_ridge
```

## ElasticNet Regression

```python
%%time
from sklearn.linear_model import ElasticNet

# Instantiate the Ridge Regressor
elastic = ElasticNet(random_state=42)

# Creating Parameter grid for GridSearch CV
params_elastic = {
    'alpha': [1e-15, 1e-10, 1e-8, 9e-4, 7e-4, 5e-4, 3e-4, 1e-4, 1e-3, 5e-2, 1e-2, 0.1, 0.3, 1, 3, 5, 10, 15, 18, 20, 30, 50, 75, 100],
    'l1_ratio': [0.001,0.01,0.1,0.3,0.5,0.8,0.9]
}

# Instantiate grid search using GradientBoostingRegressor
grid_elastic = GridSearchCV(estimator = elastic, # regressor we want to use
                       param_grid = params_elastic, # Hyperparameter space
                       scoring ='neg_mean_squared_error',
                       cv = 3, # No of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_elastic.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best ElasticNet model: {:.2f}'.format(cv_score(grid_elastic.best_score_)))
```

```python
# Store the best model 
best_model_elastic = grid_elastic.best_estimator_
```

```python
best_model_elastic
```

## SVM

```python
%%time
from sklearn.svm import SVR

# Instantiate the SVM Regressor
svr = SVR()

# Creating Parameter grid for GridSearch CV
params_svr = {
    'kernel': ['linear'], # we want a polynomial kernel
    'degree': [5,8], # degrees to test
    'gamma':[0.01,0.1], 
    'epsilon': [0.1,0.005], 
    'C': [1,15], # Penalty parameter
    'tol':[1e-7,1e-5] # Tolerance for stopping
}

# Instantiate grid search using GradientBoostingRegressor
grid_svr = GridSearchCV(estimator = svr, # regressor we want to use
                       param_grid = params_svr, # Hyperparameter space
                       scoring ='neg_mean_squared_error',
                       cv = 3, # No. of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_svr.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best SVM model: {:.2f}'.format(cv_score(grid_svr.best_score_)))
```

```python
# Store the best model 
best_model_svr = grid_svr.best_estimator_
```

```python
best_model_svr
```

## XGBoost

```python
%%time
import xgboost as xgb

# Instantiate the SVM Regressor
xgbr = xgb.XGBRegressor(learning_rate=0.01,objective='reg:linear',booster='gbtree')

# Creating Parameter grid for GridSearch CV
params_xgb = {
    'n_estimators': [8000,10000], #4000,12000
    'max_depth': [2],
    'gamma':[0.1,0.2], # Minimum loss reduction to create new tree split ,0.5,0.9
    'subsample':[0.7], 
    'reg_lambda':[0.1], 
    'reg_alpha':[0.1,0.8] 
}

# Instantiate grid search using GradientBoostingRegressor
grid_xgb = GridSearchCV(estimator = xgbr, # regressor we want to use
                       param_grid = params_xgb, # Hyperparameter space
                       scoring ='neg_mean_squared_error',
                       cv = 3, # No. of folds
                       verbose = 1,
                       n_jobs = -1) # use all cores

# fit the model
grid_xgb.fit(gaming_clean,gaming_labels)
```

```python
print('CV Score for best XGBRegressor model: {:.2f}'.format(cv_score(grid_xgb.best_score_)))
```

```python
# Store the best model 
best_model_xgb = grid_xgb.best_estimator_
```

```python
best_model_xgb
```

## Ensemble

```python
from sklearn.ensemble import VotingRegressor

voting_reg = VotingRegressor(
    estimators=[('rf', best_model_RF), ('gbr', best_model_GBR), ('elastic', best_model_elastic),
               ('ridge',best_model_ridge),('xgb',best_model_xgb)])
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(voting_reg, gaming_clean,gaming_labels,
                         scoring="neg_mean_squared_error", cv=10)
```

```python
voting_reg_score = np.sqrt(-scores)
print(voting_reg_score.mean())
```

## Predicting
<a id='PR'> </a>

```python
# Fit the Ensemble model to the training data
voting_reg.fit(gaming_clean, gaming_labels)

# Predict on test data
pred = voting_reg.predict(test_prepared)
```
